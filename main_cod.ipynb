```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(98765)
```

```python
# bsm data set
data = pd.read_csv('data/cod.csv')

col_names = data.columns
data = data.to_numpy()

# bsm
# 1 Qin ; 5 Ssi ; 7 XBH2 ;8 SNO2 ;25 SO5
covariate_col = ['1', '5', '7','25']
response_col = '30' # COD

X = data[:,np.where([t in covariate_col for t in col_names])[0]]
Y = data[:,np.where(col_names==response_col)[0]].flatten()
X = X.astype('float64')
Y = Y.astype('float64')

N = X.shape[0]
train_lag = 500 # start predicting after train_lag many observations
```

```python
%%script false

data = pd.read_csv('data/electricity-normalized.csv')
col_names = data.columns
data = data.to_numpy()

# remove the first stretch of time where 'transfer' does not vary
data = data[17760:]

# set up variables for the task (predicting 'transfer')
covariate_col = ['nswprice', 'nswdemand', 'vicprice', 'vicdemand']
response_col = 'transfer'
# keep data points for 9:00am - 12:00pm
keep_rows = np.where((data[:,2]>data[17,2])&(data[:,2]<data[24,2]))[0]

X = data[keep_rows][:,np.where([t in covariate_col for t in col_names])[0]]
Y = data[keep_rows][:,np.where(col_names==response_col)[0]].flatten()
X = X.astype('float64')[:1500, :]
Y = Y.astype('float64')[:1500]

N = X.shape[0]
train_lag = 500  # start predicting after train_lag many observations
```

```python
# run all methods
from models.cp import cqr, aecp, aecp_aci, secp, secp_aci, adaSecp, adaSecp_aci, calculate_wis

methods = [
    "cqr",
    "aecp",
    "secp",
    "adaSecp",
    "aecp_aci",
    "secp_aci",
    "adaSecp_aci",
]

# Initialize 
p_rf = np.zeros((N))
alphas = [0.4, 0.3, 0.2, 0.1, 0.05, 0.01]

# Initializes the result dictionary
interval_results = {alpha: {method: {'PI': np.zeros((N, 2)), 'cov': [], 'width': [], 'wis': 0} for method in methods} for alpha in alphas}

for alpha_val in alphas:
    print(f"Running for alpha = {alpha_val}")

    alpha_t = {method: np.zeros((N+1-train_lag)) for method in methods}
    erro_t = {method: np.zeros((N+1-train_lag)) for method in methods}
    count = 0

    for n in np.arange(train_lag, N):

        X_train = X[(n-train_lag):n, :]
        Y_train = Y[(n-train_lag):n]
        X_test = X[n:(n+1)]
        Y_test = Y[n:(n+1)]

        # train
        interval_results[alpha_val]['cqr']['PI'][n, :] = cqr(X_train, Y_train, X_test, alpha_val)
        interval_results[alpha_val]['aecp']['PI'][n, :], p_rf[n] = aecp(X_train, Y_train, X_test, alpha_val)
        interval_results[alpha_val]['secp']['PI'][n, :], p_rf[n] = secp(X_train, Y_train, X_test, alpha_val)
        interval_results[alpha_val]['adaSecp']['PI'][n, :], p_rf[n] = adaSecp(X_train, Y_train, X_test, alpha_val)
        interval_results[alpha_val]['aecp_aci']['PI'][n, :], alpha_t['aecp_aci'][count+1], erro_t['aecp_aci'][count] = aecp_aci(
            count, X_train, Y_train, X_test, Y_test, alpha_val, alpha_t['aecp_aci'][count], erro_t['aecp_aci'], gamma=0.01)
        interval_results[alpha_val]['secp_aci']['PI'][n, :], alpha_t['secp_aci'][count+1], erro_t['secp_aci'][count] = secp_aci(
            count, X_train, Y_train, X_test, Y_test, alpha_val, alpha_t['secp_aci'][count], erro_t['secp_aci'], gamma=0.01)
        interval_results[alpha_val]['adaSecp_aci']['PI'][n, :], alpha_t['adaSecp_aci'][count+1], erro_t['adaSecp_aci'][count], p_rf[n] = adaSecp_aci(
            count, X_train, Y_train, X_test, Y_test, alpha_val, alpha_t['adaSecp_aci'][count], erro_t['adaSecp_aci'], lamda=10, window_size_k=5, gamma=0.1)  # λ, k, γ

        count += 1
```

```python
# Calculate coverage, width, WIS
# print results
print("\nFinal Results for all alphas:")
print(['      ', 'coverage', 'mean width','mean wis'])
for alpha_val in alphas:
    print(f"\nResults for alpha = {alpha_val}:")
    for method in methods:
        pi_scp = interval_results[alpha_val][method]['PI']
        interval_results[alpha_val][method]['cov'] = (pi_scp[train_lag:, 0] <= Y[train_lag:]) & (pi_scp[train_lag:, 1] >= Y[train_lag:])
        interval_results[alpha_val][method]['width'] = pi_scp[train_lag:, 1] - pi_scp[train_lag:, 0]
        interval_results[alpha_val][method]['wis'] = calculate_wis(Y[train_lag:], pi_scp[train_lag:, 0], pi_scp[train_lag:, 1], alpha_val)
        
        coverage = np.mean(interval_results[alpha_val][method]['cov'])
        mean_width = np.mean(interval_results[alpha_val][method]['width'])
        mean_wis = interval_results[alpha_val][method]['wis']
        print([method, f"{coverage:.3f}", f"{mean_width:.3f}", f"{mean_wis:.3f}"])
```

```python
# box plot
import seaborn as sns
data_for_plot = {
    'Method': [],
    'Width': [],
    'CI': []
}

for alpha_val in alphas:
    for method in methods:
        widths = interval_results[alpha_val][method]['width']
        data_for_plot['Method'].extend([method] * len(widths))
        data_for_plot['Width'].extend(widths)
        data_for_plot['CI'].extend([f"{int((1-alpha_val) * 100)}%" for _ in range(len(widths))])

df_plot = pd.DataFrame(data_for_plot)

plt.figure(figsize=(12, 3))
sns.boxplot(x='CI', y='Width', hue='Method', data=df_plot)
plt.title('Prediction Interval Widths Across Methods and Confidence Intervals')
plt.ylabel('Width')
plt.xlabel('CI')
plt.savefig('./fig/pi_boxplot', dpi=500, bbox_inches='tight')
plt.show()
```
